```python
import numpy as np
```

```python
x = np.array([[100,2], [200,3], [150,2]])
y = np.array([30, 50, 40])

w = np.zeros(x.shape[1])
b = 0.0
loss_history = []
loss = 0
epochs = 1000
m = x.shape[0]
learning_rate = 0.00001
```

```python
for epoch in range(epochs):
  y_pred = np.dot(x, w) +b
  loss = np.mean((y_pred-y)**2)
  loss_history.append(loss)

  dw = np.mean((2/m) * np.dot(x.T, (y_pred - y)))
  db = np.sum((2/m) * (y_pred - y))

  w = w - learning_rate * dw
  b = b - learning_rate * db
  if epoch % 100 == 0:
    print(f"Epoch: {epoch}, Loss: {loss}")
```

```python
new = np.array([180, 2])
y_pred_new = np.dot(new, w) + b
print("y_pred_new:", y_pred_new)
```
